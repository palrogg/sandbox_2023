### Exercice, partie 1

Ce tableau de données n'est pas optimal. Comment l'améliorer? Y a-t-il des colonnes inutiles ou des valeurs à nettoyer?

```python
# Renommer une ou des colonnes provisoirement
df.rename(columns={
    "Groupe / Genre de véhicule": "genre",
    "Canton": "canton"
})
```

```python
# Renommer définitivement des colonnes
df.rename(columns={
    "Groupe / Genre de véhicule": "genre",
    "Canton": "canton"
}, inplace=True)
```

```python
# Remplacer du texte dans une colonne
# Il faudrait le faire pour supprimer les "... " des
# catégories de remorque
df["genre"].str.replace("que", "")
```

Qu'est-ce que vous pouvez trouver dans ce jeu de données?

Par exemple:
* dans quel canton le plus de remorques agricoles ont été mises en circulation en 2023?

Plus complexe:
* dans quel canton on met en circulation le plus de remorques agricoles en moyenne annuelle de 2019 à 2023? Est-ce correct de faire ce calcul?
* dans quel canton on met en circulation le plus de caravanes en proportion des autres remorques?

Vous trouvez un exemple pour les remorques agricoles en 2023 ainsi qu'un exemple de moyenne annuelle ci-dessous.

```python
# Remorques agricoles à Berne - à modifier si vous avez corrigé
# la colonne "genre"
df[df["genre"] == "... Remorque agricole"].sort_values(
    "2023",
    ascending=False
).head()
```
