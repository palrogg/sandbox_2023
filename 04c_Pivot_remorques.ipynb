```python
import pandas as pd
```

```python
# On doit préciser l'encodage, parce que l'OFS n'utilise pas le format standart
df = pd.read_csv("data/ofs-remorques.csv", encoding="Latin-1")
df.head()
```

```python
# On renomme les colonnes
df.rename(columns={
    "Groupe / Genre de véhicule": "genre",
    "Canton": "canton"
}, inplace=True)

# Et on corrige cette horreur
df["genre"] = df["genre"].str.replace("... ", "")
```

## Mettons qu’on s’intéresse à 2022 uniquement

```python
# Pas de calcul: on pivote les données sans les modifier
pd.pivot(df, index="genre", columns="canton", values="2022")
```

```python
# Mettons qu’on veut additionner les catégories de remorques
pd.pivot_table(df, index="canton", values=["2021", "2022"], aggfunc="sum")
```
